# **Analisis y visualizacion de el tamaño de las aletas en el conjunto de datos de los pinguinos**

##Importar librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Cargar archivo y mostrarlo

```python
df = pd.read_csv('pinguinos.csv')
df
```

## Limpieza de datos

```python
df = df.dropna(subset = ['flipper_length_mm'], axis ='index')
df
```

## Grafico de cajas

```python
df.boxplot(column = 'flipper_length_mm', by = 'sex')
plt.show()
```

## Grafico seaborn

```python
sns.boxplot(df, x = 'sex', y = 'flipper_length_mm').set(title = 'Agrupacion de tamaño de aletas por sexo')
plt.show()
```

## Grafica de violin

```python
sns.violinplot(df, x = 'sex', y = 'flipper_length_mm').set(title = 'Agrupacion de tamaño de aletas por sexo')
plt.show()
```
